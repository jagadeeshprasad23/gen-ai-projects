```python
from langchain.llms import OpenAI
from langchain.retrievers.self_query.base import SelfQueryRetriever
from langchain.chains.query_constructor.base import AttributeInfo

metadata_field_info = [
    AttributeInfo(
        name="ID",
        description="It is the product ID, a unique identifier for each product",
        type="string",
    ),
    AttributeInfo(
        name="price",
        description="The price of the product in GBP",
        type="float",
    ),
    AttributeInfo(
        name="weight",
        description="The weight of the product in grams or other units, or 'unknown' if not provided",
        type="string",
    ),
    AttributeInfo(
        name="package",
        description="The type of packaging of the product, such as 'tin', 'packet', or 'box'",
        type="string",
    ),
    AttributeInfo(
        name="tags",
        description="A list of tags describing the product(coma seperated), such as 'vegan', 'vegetarian', or 'halal'",
        type="string",
    ),
    AttributeInfo(
        name="free_from",
        description="A list of allergens or substances the product is free from(coma seperated), such as 'gluten', 'egg', or 'dairy'",
        type="string",
    ),
]

document_content_description = (
    "Description of the Document, including the product details like name, ID, price, weight, package type, "
    "tags for dietary preferences, and allergens or substances the product is free from"
)

llm = OpenAI(openai_api_key=openai_api_key, temperature=0)

retriever = SelfQueryRetriever.from_llm(
    llm,
    vectorstore,
    document_content_description,
    metadata_field_info,
    enable_limit=True,  # To limit the returning documents count
    verbose=True,
)
```

```python
# Get the relevant documents
relevant_docs = retriever.get_relevant_documents("Tomatoes in tins")

# Iterate over the documents and print the required fields
for doc in relevant_docs:
    print(f"ID: {doc.metadata['ID']} , Name: {doc.page_content}, Price: {doc.metadata['price']}")
```
